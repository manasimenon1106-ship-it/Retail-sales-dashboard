```python
yearly_sales = sales.groupby('Year')['Sales'].sum()

print(yearly_sales)
```

```python
yoy_growth = yearly_sales.pct_change() * 100

print(yoy_growth)
```

```python
yoy_table = yearly_sales.to_frame(name='Sales')

yoy_table['YoY Growth %'] = yoy_growth

print(yoy_table)
```

```python
import matplotlib.pyplot as plt

yoy_growth.plot(marker='o', figsize=(8,5))

plt.title('Year-over-Year Sales Growth')
plt.xlabel('Year')
plt.ylabel('Growth (%)')
plt.grid(True)

plt.show()
```

```python
customer_orders = sales.groupby('CustomerID')['OrderID'].nunique()

repeat_customers = (customer_orders > 1).sum()

repeat_customer_rate = (
    repeat_customers /
    customer_orders.count()
) * 100

print("Repeat Customer Rate (%):", round(repeat_customer_rate,2))
```
